```python
def draw_chessboard(chessboard):
    # print the chessboard with the pieces in their positions
    print("")
    for row in range(8, 0, -1):
        # print the row number on the left side of the board
        print(row, end=" ")
        for col in range(97, 105):
            # calculate the square name based on the row and column numbers
            square = chr(col) + str(row)
            # check if there is a piece on the square
            piece = chessboard.get(square, "")
            # print the piece or an empty square
            if piece=="":
                print("| {}  ".format(piece), end="")
            else:
                print("| {} ".format(piece), end="")
        # print the row number on the right side of the board
        print("|", row)
        for col in range(97, 105):
            print("-----",end="")
        print("\n")
        # print the column letters on the bottom of the board
    print("   a   b   c   d   e   f   g   h")
```

```python
def available_pieces(chessboard):
    available=[]
    my_pieces = {"K", "Q", "R", "B", "N", "P"}
    for square,piece in chessboard.items():
        if piece in my_pieces:
            available.append(square)
    return available
    
```

```python
def available_pieces_AI(chessboard):
    available=[]
    AI_pieces = {"k", "q", "r", "b", "n", "p"}
    for square,piece in chessboard.items():
        if piece in AI_pieces:
            available.append(square)
    return available
    
```

```python
def available_moves(chessboard, my_piece, position, AI):
    available = []
    if AI == False:
        opp_pieces = {"k", "q", "r", "b", "n", "p"}
        my_pieces = {"K", "Q", "R", "B", "N", "P"}
    else:
        my_pieces = {"k", "q", "r", "b", "n", "p"}
        opp_pieces = {"K", "Q", "R", "B", "N", "P"}

    def is_path_clear(start, end):
        if start[0] == end[0]:
            # Check if any squares between start and end have a piece in the row
            for i in range(min(int(start[1]), int(end[1])) + 1, max(int(start[1]), int(end[1]))):
                if chessboard[start[0] + str(i)] != "":
                    return False
        elif start[1] == end[1]:
            # Check if any squares between start and end have a piece in the coloumn
            for i in range(ord(min(start[0], end[0])) + 1, ord(max(start[0], end[0]))):
                if chessboard[chr(i) + start[1]] != "":
                    return False
        elif abs(ord(start[0]) - ord(end[0])) == abs(int(start[1]) - int(end[1])):
            # Check if any squares between start and end have a piece 
            col_diff = 1 if start[0] < end[0] else -1
            row_diff = 1 if start[1] < end[1] else -1
            curr_col = ord(start[0]) + col_diff
            curr_row = int(start[1]) + row_diff
            while curr_col != ord(end[0]) and curr_row != int(end[1]):
                if chessboard[chr(curr_col) + str(curr_row)] != "":
                    return False
                curr_col += col_diff
                curr_row += row_diff
        return True

    for square, piece in chessboard.items():
        if (piece == "" or (piece in opp_pieces)):
            if my_piece == "R" or my_piece == "r":
                if (square[0] == position[0] or square[1] == position[1]) and is_path_clear(position, square):
                    if(position!=square):
                        available.append(square)
            elif my_piece == "P" or my_piece == "p":
                if position[1] == "2" and my_piece == "P":
                    if (square[1] == "3" or square[1] == "4") and (square[0] == position[0]) and (chessboard[square[0] + "3"] == "" and chessboard[square[0] + "4"] == ""):
                        if(position!=square):
                            available.append(square)
                elif position[1] != "2" and my_piece == "P":
                    diff = int(square[1]) - int(position[1])
                    if (square[1] > position[1] and diff == 1) and (square[0] == position[0]) and (chessboard[square[0] + str(int(position[1])+1)]) in opp_pieces:
                        if(position!=square):
                            available.append(square)
                elif position[1] == "7" and my_piece == "p":
                    if (square[1] == "6" or square[1] == "5") and (square[0] == position[0]) and (chessboard[square[0] + "6"] == "" and chessboard[square[0] + "5"] == ""):
                        if(position!=square):
                            available.append(square)
                elif position[1] != "7" and my_piece == "p":
                    diff = int(position[1]) - int(square[1])
                    if (square[1] < position[1] and diff == 1) and (square[0] == position[0]) and (chessboard[square[0] + str(int(position[1])-1)] in opp_pieces):
                        if(position!=square):
                            available.append(square)
            elif my_piece == "N" or my_piece == "n":
                col_diff = abs(ord(position[0]) - ord(square[0]))
                row_diff = abs(int(position[1]) - int(square[1]))
                if (col_diff == 1 and row_diff == 2) or (col_diff == 2 and row_diff == 1):
                    if(position!=square):
                        available.append(square)
            elif my_piece == "B" or my_piece == "b":
                if abs(ord(position[0]) - ord(square[0])) == abs(int(position[1]) - int(square[1])) and is_path_clear(position, square):
                    if(position!=square):
                        available.append(square)
            elif my_piece == "Q" or my_piece == "q":
                if (square[0] == position[0]) or (square[1] == position[1]) and is_path_clear(position, square):
                    if(position!=square):
                        available.append(square)
                elif abs(ord(position[0]) - ord(square[0])) == abs(int(position[1]) - int(square[1])) and is_path_clear(position, square):
                    if(position!=square):
                        available.append(square)
            elif my_piece == "K" or my_piece == "k":
                col_diff = abs(ord(position[0]) - ord(square[0]))
                row_diff = abs(int(position[1]) - int(square[1]))
                if (col_diff == 1 and row_diff == 1) or (col_diff == 1 and row_diff == 0) or (col_diff == 0 and row_diff == 1):
                    if(position!=square):
                        available.append(square)
    return available

```

```python
def fitness(chessboard):
    piece_values = {"P": 1, "N": 3, "B": 3, "R": 5, "Q": 9, "K": 0}
    position_values = {
        "P": [
            [0,  0,  0,  0,  0,  0,  0,  0],
            [50, 50, 50, 50, 50, 50, 50, 50],
            [10, 10, 20, 30, 30, 20, 10, 10],
            [5,  5, 10, 25, 25, 10,  5,  5],
            [0,  0,  0, 20, 20,  0,  0,  0],
            [5, -5,-10,  0,  0,-10, -5,  5],
            [5, 10, 10,-20,-20, 10, 10,  5],
            [0,  0,  0,  0,  0,  0,  0,  0]
        ],
        "N": [
            [-50,-40,-30,-30,-30,-30,-40,-50],
            [-40,-20,  0,  0,  0,  0,-20,-40],
            [-30,  0, 10, 15, 15, 10,  0,-30],
            [-30,  5, 15, 20, 20, 15,  5,-30],
            [-30,  0, 15, 20, 20, 15,  0,-30],
            [-30,  5, 10, 15, 15, 10,  5,-30],
            [-40,-20,  0,  5,  5,  0,-20,-40],
            [-50,-40,-30,-30,-30,-30,-40,-50]
        ],
        "B": [
            [-20,-10,-10,-10,-10,-10,-10,-20],
            [-10,  0,  0,  0,  0,  0,  0,-10],
            [-10,  0,  5, 10, 10,  5,  0,-10],
            [-10,  5,  5, 10, 10,  5,  5,-10],
            [-10,  0, 10, 10, 10, 10,  0,-10],
            [-10, 10, 10, 10, 10, 10, 10,-10],
            [-10,  5,  0,  0,  0,  0,  5,-10],
            [-20,-10,-10,-10,-10,-10,-10,-20]
        ],
        "R": [
                [-5,  0,  5,  0,  0,  5,  0, -5],
                [-5,  0,  5,  0,  0,  5,  0, -5],
                [-5,  0,  5,  0,  0,  5,  0, -5],
                [-5,  0,  5,  0,  0,  5,  0, -5],
                [0,   0,  5,  5,  5,  5,  0,  0],
                [5,  10, 10, 10, 10, 10, 10,  5],
                [0,   0,  5,  0,  0,  0,  0,  0],
                [-20,-10,-10, -5, -5,-10,-10,-20]
                ],
    "Q": [
        [-20,-10,-10, -5, -5,-10,-10,-20],
        [-10,  0,  0,  0,  0,  0,  0,-10],
        [-10,  0,  5,  5,  5,  5,  0,-10],
        [-5,   0,  5,  5,  5,  5,  0, -5],
        [0,    0,  5,  5,  5,  5,  0, -5],
        [-10,  5,  5,  5,  5,  5,  0,-10],
        [-10,  0,  5,  0,  0,  0,  0,-10],
        [-20,-10,-10, -5, -5,-10,-10,-20]
    ]
    }

    total_value = 0
    for square,piece in chessboard.items():
            #piece = chessboard[(row, col)]
            if piece != "":
                if piece.upper() !="K":
                    if piece.isupper():
                        break
                    else:
                        color = "B" if piece.isupper() else "W"
                        value = piece_values[piece.upper()]
                        if color == "W":
                            value *= -1
                        values=position_values.pop(piece.upper())
                        position_values[piece.upper()]=values
                        for row in values:
                            for value in row:
                                total_value += value
                
    return total_value
```

```python
def minimax(chessboard, depth, alpha, beta, optimised_player):
    if depth == 0:
        return fitness(chessboard)
    if optimised_player:
        max_eval = -float('inf')
        for initial in available_pieces(chessboard):
            
            piece=chessboard.pop(initial)
            chessboard[initial]=""
            for new in available_moves(chessboard, piece, initial,False):
                #make the move
                #print("new: ",new)
                removed_piece=chessboard.pop(new)
                chessboard[new]=piece
                eval = minimax(chessboard, depth-1, alpha, beta, False)
                #un-make the move
                #chessboard[initial]=piece
                chessboard[new]=removed_piece
                max_eval = max(max_eval, eval)
                alpha = max(alpha, eval)
                if beta <= alpha:
                    chessboard[initial]=piece
                    break
            chessboard[initial]=piece
        return max_eval
    else:
        min_eval = float('inf')
        for initial in available_pieces(chessboard):
            piece=chessboard.pop(initial)
            #chessboard[initial]=""
            for new in available_moves(chessboard, piece, initial,False):
                removed_piece=chessboard.pop(new)
                chessboard[new]=piece
                eval = minimax(chessboard, depth-1, alpha, beta, True)
                chessboard[new]=removed_piece
                min_eval = min(min_eval, eval)
                beta = min(beta, eval)
                if beta <= alpha:
                    chessboard[initial]=piece
                    break
            chessboard[initial]=piece
        return min_eval
```
